```python

df_np = np.array(df)


# Column indices for major and non-major traffic
major_traffic_idx = 3
non_major_traffic_idx = 4

# total traffic handled by major and non-major ports
total_major_traffic = np.nansum(df_np[:, major_traffic_idx].astype(float))
total_non_major_traffic = np.nansum(df_np[:, non_major_traffic_idx].astype(float))

#  the average traffic handled by major and non-major ports
avg_major_traffic = np.nanmean(df_np[:, major_traffic_idx].astype(float))
avg_non_major_traffic = np.nanmean(df_np[:, non_major_traffic_idx].astype(float))

# Print results
print(f"Total traffic handled by major ports: {total_major_traffic} tonnes")
print(f"Total traffic handled by non-major ports: {total_non_major_traffic} tonnes")
print(f"Average traffic handled by major ports: {avg_major_traffic} tonnes")
print(f"Average traffic handled by non-major ports: {avg_non_major_traffic} tonnes")
```

```python
 #Line Plot for Yearly Traffic Trend for Major Ports
import seaborn as sns # Import the seaborn library
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Traffic Handled By Major Ports (UOM:t(Tonnes))', hue='State', data=df, marker='o')
plt.title('Yearly Traffic Trend for Major Ports')
plt.xlabel('Year')
plt.ylabel('Traffic Handled (Tonnes)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 4. Line Plot for Yearly Traffic Trend for Non-Major Ports
import seaborn as sns # Import the seaborn library
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Traffic Handled By Non Major Ports (UOM:t(Tonnes))', hue='State', data=df, marker='o')
plt.title('Yearly Traffic Trend for Non-Major Ports')
plt.xlabel('Year')
plt.ylabel('Traffic Handled (Tonnes)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

df_np = np.array(df)

states = np.unique(df_np[:, 1])

total_major_traffic = []
total_non_major_traffic = []

for state in states:
    state_data = df_np[df_np[:, 1] == state]
    total_major_traffic.append(np.nansum(state_data[:, 3].astype(float)))
    total_non_major_traffic.append(np.nansum(state_data[:, 4].astype(float)))

plt.figure(figsize=(12, 8))
bar_width = 0.4
index = np.arange(len(states))

plt.bar(index, total_major_traffic, width=bar_width, label='Major Ports')
plt.bar(index + bar_width, total_non_major_traffic, width=bar_width, label='Non-Major Ports')

plt.xlabel('States')
plt.ylabel('Total Traffic (Tonnes)')
plt.title('Total Traffic Handled by Major and Non-Major Ports')
plt.xticks(index + bar_width/2, states, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Major Ports
traffic_major_ports = df.groupby('State')['Traffic Handled By Major Ports (UOM:t(Tonnes))'].sum()
traffic_major_ports.plot.pie(autopct='%1.1f%%', figsize=(8, 8), title='Traffic Distribution by State (Major Ports)')
plt.ylabel('')
plt.show()
```

```python
# Non Major Ports
traffic_non_major_ports = df.groupby('State')['Traffic Handled By Non Major Ports (UOM:t(Tonnes))'].sum()
traffic_non_major_ports.plot.pie(autopct='%1.1f%%', figsize=(8, 8), title='Traffic Distribution by State (Non-Major Ports)')
plt.ylabel('')
plt.show()
```

```python

stacked_data = df.groupby(['State', 'Year'])[['Traffic Handled By Major Ports (UOM:t(Tonnes))',
                                              'Traffic Handled By Non Major Ports (UOM:t(Tonnes))']].sum().reset_index()


stacked_data_melted = pd.melt(stacked_data, id_vars=['State', 'Year'],
                              value_vars=['Traffic Handled By Major Ports (UOM:t(Tonnes))',
                                          'Traffic Handled By Non Major Ports (UOM:t(Tonnes))'],
                              var_name='Port Type', value_name='Traffic (Tonnes)')
plt.figure(figsize=(12, 8))
sns.barplot(x='Year', y='Traffic (Tonnes)', hue='State',
            data=stacked_data_melted, palette='viridis', ci=None,
            hue_order=stacked_data['State'].unique())
plt.title('Contribution of Major and Non-Major Ports to Total Traffic')
plt.xlabel('Financial Year')
plt.ylabel('Traffic (Tonnes)')
plt.xticks(rotation=45)
plt.legend(title='State', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

Analysis

Gujarat's major ports handled a steadily increasing volume of traffic. the analysis reveals that major ports in states like Andhra Pradesh, Gujarat, and Maharashtra play a crucial role in India's port traffic, handling significant volumes over the years. Non-major ports, while handling lesser traffic overall, are essential in supporting the major ports and accommodating regional and smaller-scale trade activities.
